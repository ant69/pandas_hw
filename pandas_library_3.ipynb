## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

___
## Решение задания 1

```python
import pandas as pd
# Определяем функцию для вычисления значения новой колонки
def get_type(row):
    if row['traffic_source'] in ['yandex', 'google']:
        ret = 'organic'
    elif row['traffic_source'] in ['pain', 'email']:
        if row['region'] == 'Russia':
            ret = 'ad'
        else:
            ret = 'other'
    else:
        ret = row['traffic_source']
    return ret

# Читаем файл
log_ds = pd.read_csv('visit_log.csv', engine='python', sep=';')
# Создаем новую колонку и заполняем значения в ней для каждой строки
log_ds['source_type'] = log_ds.apply(get_type, axis=1)
log_ds.head(20)
```

## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

___
## Решение задания 3

```python
import pandas as pd

# Определяем функцию для вычисления времени жизни пользователей
def get_livetime(d):
    return d.max() - d.min()

# Загружаем файл ratings.csv с рейтингами фильмов
ratings_dataset = pd.read_csv('ml-latest-small/ratings.csv', engine='python')
# после чтения датасета группируем его по пользователям, подсчитывая количество оценок и время жизни
grouped_rating = ratings_dataset.groupby('userId').agg( {'timestamp': ['count', get_livetime]} ).reset_index()
# фильтруем рейтинг, удаляя из него пользователей, давших менее 100 оценок
grouped_rating[ ( grouped_rating['timestamp']['count'] > 99 )].head(10)
```

## Задание 4

Дана статистика услуг перевозок клиентов компании по типам:

rzd - железнодорожные перевозки
auto - автомобильные перевозки
air - воздушные перевозки
client_base - адреса клиентов

Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

___
## Решение задания 4

```python
import pandas as pd
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# Создаем Dataframe clients_revenues, последовательно объединяя три датафрейма с ревеню по клиентам
# Используем outer join, чтобы не потерять данные по отдельным клиентам
clients_revenues = pd.merge(rzd, auto, on='client_id',  how='outer')
clients_revenues = pd.merge(clients_revenues, air, on='client_id', how='outer')
clients_revenues
```

```python
# Создаем датафрейм с адресами клиентов, объединяя с датафреймом клиентов ранее созданный датафрейм с их ревеню
full_clients = pd.merge(client_base, clients_revenues, on='client_id', how='left')
full_clients
```
